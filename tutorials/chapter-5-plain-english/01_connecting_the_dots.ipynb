## Start with something you already know

If you've ever drawn a family tree, you've made a knowledge graph.

- You drew little circles or boxes for each person.
- You drew lines between them to show how they're related (`is the parent of`, `is married to`, `is the sibling of`).

That's literally it. **A graph is a bunch of dots connected by lines.** A *knowledge graph* is the same thing, but the dots stand for real things in the world (people, places, products, ideas) and the lines stand for real relationships between them.

If you've seen one of these in the wild, you've probably seen it in:

- **Google's search results** — when you search for a movie and a panel pops up on the right with the director, cast, year, and budget.
- **Netflix or Spotify recommendations** — "Because you watched X, you might like Y" is a graph traversal at heart.
- **LinkedIn's "People You May Know"** — built directly on top of a graph of who-knows-who.

The trick in this book is teaching the search engine to *build one of these graphs automatically* — from the content it already has — and then use it to give you smarter search results.

## The two pieces of every graph

Every graph in the world is made of just two things.

### 🟢 Nodes

Also called **vertices** (because mathematicians can't help themselves). A node is *a thing*. A person. A place. A movie. A word. A concept. It doesn't matter what — if it can be named, it can be a node.

### ➡️ Edges

An edge is *a relationship between two things*. It connects two nodes and (usually) has a label that says how they're related.

> **🎯 Remember.** **Node = thing. Edge = how two things are connected.** That's the whole vocabulary. You can talk to a graph database engineer with just these two words.

Here's a tiny example. Imagine three nodes — `Lucy`, `Linus`, `Charlie Brown` — and the edges between them:

- `Lucy` `is the sibling of` `Linus`
- `Lucy` `is friends with` `Charlie Brown`
- `Linus` `is friends with` `Charlie Brown`

If you drew that on a napkin, you'd already have a knowledge graph. Let's do exactly that, in code.

## A graph you can actually see

The code in the next cell uses a Python library called `networkx` to build a tiny graph in memory, and `matplotlib` to draw a picture of it. If you've never run Python before, that's fine — just click the ▶️ button to the left of the cell. (In Jupyter, you can also press Shift+Enter.)

You shouldn't need to understand the code to understand the picture it makes.

````python
import networkx as nx
import matplotlib.pyplot as plt

# 1. Make an empty graph.
G = nx.Graph()

# 2. Add three nodes (the "things").
G.add_node("Lucy")
G.add_node("Linus")
G.add_node("Charlie Brown")

# 3. Add edges (the "relationships").
G.add_edge("Lucy", "Linus", relationship="sibling of")
G.add_edge("Lucy", "Charlie Brown", relationship="friends with")
G.add_edge("Linus", "Charlie Brown", relationship="friends with")

# 4. Draw it.
fig, ax = plt.subplots(figsize=(7, 5))
pos = nx.spring_layout(G, seed=42)
nx.draw_networkx_nodes(G, pos, node_color="#FFD43B", node_size=2500, ax=ax)
nx.draw_networkx_labels(G, pos, font_size=11, font_weight="bold", ax=ax)
nx.draw_networkx_edges(G, pos, width=2, edge_color="#888", ax=ax)
edge_labels = nx.get_edge_attributes(G, "relationship")
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=9, ax=ax)
ax.set_title("A tiny knowledge graph", fontsize=13)
ax.axis("off")
plt.tight_layout()
plt.show()
````

That's a knowledge graph. Three nodes, three edges. You could expand it forever — add `Sally`, `Snoopy`, `Woodstock`, the dates of every comic strip they appeared in, the city they live in — and it would still be the same two pieces: nodes and edges.

> **💡 Tip.** When a knowledge graph is described as having "a billion edges" or "100 million nodes," don't be intimidated. It's the same Lucy/Linus/Charlie Brown idea, scaled up. The hard problem isn't drawing the graph — it's deciding what nodes and edges should exist in the first place.

## When a graph becomes a *knowledge* graph

Not every graph is a knowledge graph. A graph could be:

- A road map (nodes = intersections, edges = streets) — that's a graph, not really "knowledge."
- The wiring diagram of a microwave — also a graph, also not really "knowledge."

A graph becomes a **knowledge graph** when its nodes and edges represent *facts about the world*. The shift is fuzzy, but in practice it means:

1. The nodes stand for real-world things people can name (people, products, ideas).
2. The edges describe meaningful relationships (`is a`, `wrote`, `married to`, `located in`).
3. You can use it to answer questions you couldn't easily answer otherwise.

That third part is the punchline. The reason anyone bothers building one of these is to answer questions like:

- *"Show me drugs that work the same way as advil."*
- *"What other movies are in the Marvel Cinematic Universe?"*
- *"Who else in our company has worked with this client?"*

Keyword search alone can't answer those questions. A knowledge graph can.

## Three ways to build a knowledge graph

There are three main paths people take. The chapter walks through all three, but spends most of its time on the third one (and so will we).

### Option 1: Build it by hand 🛠️

You sit down with a domain expert and type in every node and edge yourself.

- **Pros:** Highest quality. Captures domain expertise precisely.
- **Cons:** Slow. Expensive. Becomes outdated the moment you stop maintaining it. Doesn't scale beyond hundreds or thousands of facts.
- **Used by:** Specialized industries where accuracy beats speed (medical coding, legal taxonomies).

### Option 2: Plug in an existing one 📚

Use a public knowledge graph someone else built. The big names:

- **ConceptNet** — common-sense connections between everyday concepts.
- **DBpedia / Wikidata** — structured facts pulled from Wikipedia.
- **Large language models** — increasingly, LLMs are themselves *implicit* knowledge graphs.

- **Pros:** Free. Huge. Available immediately.
- **Cons:** Generic. Won't know about *your* products, *your* customers, *your* internal jargon. 
- **Used by:** General-purpose web search, chatbots that need broad common knowledge.

### Option 3: Have your search engine build one for you 🤖

Feed the search engine your documents and let it figure out the relationships from the text itself, automatically.

- **Pros:** Free (you already have the documents). Updates itself as your content grows. Knows about *your* terminology.
- **Cons:** Quality depends on how much content you have. Won't outperform a hand-built graph in a narrow domain.
- **Used by:** Search engines that need to understand domain-specific language without an army of curators.

**This is what the rest of Chapter 5 is about.** The next four notebooks in this series build up the third option, step by step.

## A side-by-side comparison

| | **By hand** | **Plug in existing** | **Autogenerated** |
|---|---|---|---|
| **Effort to set up** | Months of curation | Hours | Hours |
| **Cost** | High (expert time) | Free or licensed | Free (uses what you have) |
| **Domain accuracy** | Excellent | Generic | Tuned to your data |
| **Updates** | Manual | Periodic | Continuous |
| **Best for** | Regulated industries | Broad / web-scale | Domain-specific apps |

> **⚠️ Warning.** These approaches aren't mutually exclusive — many real systems combine all three. The book's Chapter 5 example, for instance, autogenerates a graph from forum posts *and* lets you sprinkle in some hand-curated facts on top.

## 🤓 Technical Stuff — skippable

*Feel free to skip this section. It's here for the curious.*

When graph people talk about a fact in a knowledge graph, they often write it as a three-part thing called an **RDF triple** (Resource Description Framework — the name is uglier than the idea).

A triple has three parts:

```
subject  →  predicate  →  object
```

So `(Lucy, "is the sibling of", Linus)` is a triple. The subject is the starting node, the predicate is the edge label, and the object is the ending node.

That's the whole standard. The reason it matters is that lots of graph databases and graph file formats (Turtle, JSON-LD, RDF/XML) are essentially big lists of triples. If you see the word "triple" in graph documentation, just translate it to *(thing, relationship, thing)* and you'll be fine.

A few graph databases worth knowing the names of (so you don't get caught off guard in a meeting):

- **Neo4j** — the popular one. Uses a query language called Cypher.
- **Apache TinkerPop / Gremlin** — a graph traversal framework usable with many databases.
- **ArangoDB** — a multi-model database that also does graphs.
- **RDF triple stores** — Apache Jena, GraphDB, Stardog.

This book doesn't use any of these. It uses Apache Solr — *a search engine* — as the graph database. That sounds surprising, and it's the punchline of notebook #2 in this series.

## Cheat sheet

If you remember only three things from this notebook:

1. **A graph is dots connected by lines.** Nodes are things, edges are relationships. That's it.
2. **A knowledge graph is a graph where the dots and lines represent real-world facts** — useful for answering questions keyword search can't.
3. **There are three ways to build one** — by hand, plug in an existing one, or autogenerate from your content. This book uses the third option, with a clever twist: the search engine *itself* becomes the graph.

## Glossary

| Term | What it means in plain English |
|------|-------------------------------|
| **Node** | A thing in the graph. Also called a *vertex*. |
| **Edge** | A relationship between two nodes. Usually has a label. |
| **Vertex** | Math word for node. Don't be intimidated. |
| **Knowledge graph** | A graph where the nodes and edges represent facts about the world. |
| **Ontology** | A formal description of which kinds of nodes and edges are allowed in a domain. Think "schema for knowledge." |
| **Taxonomy** | A tree-shaped knowledge structure — everything connected by "is a kind of." A simpler cousin of a knowledge graph. |
| **RDF triple** | A fact written as `(subject, relationship, object)`. The plumbing of most knowledge-graph standards. |
| **Graph database** | A database optimized for storing nodes and edges. Examples: Neo4j, ArangoDB. |
